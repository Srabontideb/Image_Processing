* A negative of an image or inverted is computed using the following formula:<br>
    **s = (L  1)  r**<br>
Where, r is the input image pixel, L is the highest value of the pixel and s is the resulting pixel.

In the context of image processing, a Negative Transformation involves taking the original pixel values of the grayscale image and computing their negative counterparts. The transformation is achieved by subtracting each pixel's intensity value from the maximum intensity value in the grayscale, which is 255 for an 8-bit image.

### **Calculating the Negative Image:**
The code uses the formula ***negative = 255 - image*** to calculate the negative version of the loaded grayscale image. Here, image is the NumPy array representing the original grayscale image, and negative is a new NumPy array storing the negative version.

**Log Transformation**
* During log transformation, the dark pixels in an image are expanded as compare to the higher pixel values.
The log transformation of an image is computed using the following formula:



*   **c=255/log(1+maximum intensity value)**
*   **s=c*log(1+r)**






````python
# Log Transformation
c = round(255 / np.log(1 + np.max(image)))
log_image = c * (np.log(1+image))
log_image = np.array(log_image, dtype = np.uint8)
````

### **[EXPLANATION](https://)**
### **Determining the Scaling Factor:**
`c = round(255 / np.log(1 + np.max(image)))`

The code computes a scaling factor c based on the maximum pixel intensity of the original image. This scaling factor is used to stretch the pixel values during the log transformation.

### **Log Transformation:**
`log_image = c * (np.log(1 + image))`

The log transformation is applied to each pixel of the original grayscale image to create the log_image. The transformation helps to spread out the pixel values, enhancing the contrast of the image.

### **Converting to Unsigned 8-bit Integer (uint8):**
`log_image = np.array(log_image, dtype=np.uint8)`

To ensure that the pixel values are within the valid 8-bit range (0 to 255), the log_image is converted to a NumPy array of data type np.uint8.

**Gamma Correction**
* Gamma transformation or power law transformation involve nth power and nth root transformation.
* An image is gamma transformed using the following formula:
 **s = c * (r^)**
* Variation in the value of  varies the enhancement of the images.
* This type of transformation is used for enhancing images for different type of display devices.

````python
gamma_image = np.zeros_like(image)
gamma = 0.5
c = 255 / (255 ** gamma)
# Apply gamma correction using a for loop
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        pixel_value = image[i,j]
        corrected_value = c * (pixel_value ** gamma)
        gamma_image[i,j] = corrected_value
````

**Contrast Stretching**
* Low contrast image occur often due to improper illumination or non-linearly or small dynamic range of an imaging sensor.
* It increases the dynamic range of grey levels in the image.
It has three cases:
1. **When r1 =s1 and r2=s2, transformation becomes a Linear function.**
2. **When r1=r2, s1=0 and s2=L-1, transformation becomes a thresholding function.**
3. **When (r1, s1) = (rmin, 0) and (r2, s2) = (rmax, L-1), this is known as Min-Max Stretching.**

**Image Thresholding**

````python
# Importing the Libraries & Loading the Image
import cv2
import matplotlib.pyplot as plt
import numpy as np
import warnings
image = cv2.imread('/content/road.bmp', cv2.IMREAD_GRAYSCALE)
thresh_value = 128
# Create empty binary image
binary = np.zeros_like(image)
# Loop over each pixel in the image
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        # Apply thresholding
        if image[i,j] >= thresh_value:
            binary[i,j] = 255
        else:
            binary[i,j] = 0
````

3. **Min-Max Contrast Stretching**

````python
image = cv2.imread('/content/Low Contrast.PNG', cv2.IMREAD_GRAYSCALE)
r1, r2 = image.min(), image.max()
s1, s2 = 0, 255
# Apply the linear transformation using a for loop
output_image = np.zeros_like(image)
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        output_image[i,j] = (image[i,j] - r1) * ((s2 - s1) / (r2 - r1)) + s1
# Displaying Results
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax1.imshow(image, cmap='gray', vmin=0, vmax=255)
ax1.title.set_text('Original Image')
ax2.imshow(output_image, cmap='gray', vmin=0, vmax=255)
ax2.title.set_text('Contrast Stretched Image')
plt.show()
````

### **[EXPLANATION](https://)**
### **Loading the Image:**
`image = cv2.imread('/content/Low Contrast.PNG', cv2.IMREAD_GRAYSCALE)`
The code begins by loading the image "Low Contrast.PNG" into a NumPy array using OpenCV's imread() function with the cv2.IMREAD_GRAYSCALE flag, which reads the image in grayscale mode.
### **Finding the Minimum and Maximum Intensity Values:**
`r1, r2 = image.min(), image.max()`
The variables r1 and r2 are assigned the minimum and maximum pixel intensity values in the original grayscale image, respectively.
### **Defining the New Intensity Range:**
`s1, s2 = 0, 255`
The variables s1 and s2 are set to 0 and 255, respectively. These values represent the new minimum and maximum intensity values for the output contrast-stretched image.
### **Applying the Linear Transformation:**
```
output_image = np.zeros_like(image)
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        output_image[i, j] = (image[i, j] - r1) * ((s2 - s1) / (r2 - r1)) + s1

```
The nested loops iterate through each pixel of the original grayscale image. For each pixel at position (i, j), a linear transformation is applied to map the pixel intensity from the range [r1, r2] to the range [s1, s2]
### **Storing Contrast Stretched Pixel Values:**
The calculated pixel value after the linear transformation is stored in the corresponding position (i, j) of the output_image array.

### **Displaying Results:**
```
# Displaying Results
```

**Intensity Level Slicing**
* Intensity level slicing involves highlighting a specific range of grey level in an image.

* **Case-I:**
1. To display a high value for all gray levels in the range of interest.
2. To display a low value for the rest of gray levels.

* **Case-II:**
1. Brighten the desired range of gray levels.
2. Preserve the rest of the intensity levels remain unchanged as same as the original gray levels.

````python
image = cv2.imread('/content/Dark.PNG', cv2.IMREAD_GRAYSCALE)
# Parameters
minimum_range = 64
maximum_range = 128
output_image = np.zeros_like(image)
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        if image[i,j] > minimum_range and image[i,j] < maximum_range:
            output_image[i,j] = 255
        else:
            output_image[i,j] = 0 # image[i,j] for second case
# Displaying Results
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

ax1.imshow(image, cmap='gray', vmin=0, vmax=255)
ax1.title.set_text('Original Image')
ax2.imshow(output_image, cmap='gray', vmin=0, vmax=255)
ax2.title.set_text(f'Intensity Level Sliced Image\nMin:{minimum_range}, Max:{maximum_range}')
plt.show()
````
