```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
img = cv2.imread('Dark.PNG')
```

```python
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Flatten the image to a 1D array
gray_flat = gray.ravel()
```

```python
# Calculate the histogram
hist = cv2.calcHist([gray_flat], [0], None, [256], [0, 256])
```

```python
# Plot the histogram
plt.subplot(2,1,1).set_title('original image')
plt.plot(hist,color='green')
plt.subplot(2,1,2)
plt.imshow(img)
plt.show()
```

```python
# Apply histogram equalization
eq = cv2.equalizeHist(gray)
hist2 = cv2.calcHist([eq], [0], None, [256], [0, 256])
plt.subplot(2,1,1).set_title('equalized image')
plt.plot(hist2,color='green')
plt.subplot(2,1,2)
plt.imshow(eq,cmap='gray')
plt.show()
```
